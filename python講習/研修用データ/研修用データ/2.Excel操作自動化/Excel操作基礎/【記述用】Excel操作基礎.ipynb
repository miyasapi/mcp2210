# PythonからExcelを操作する　～基本編

***

openpyxlのインポート

```python
import openpyxl
```

## １．ワークブックとワークシート

■ワークブックの新規作成

```python
wb = openpyxl.Workbook()
```

■対象シートの絞り込み（シート名）

```python
ws = wb['Sheet']
print(ws)
```

■ワークブックの読み込み

```python
wb2 = openpyxl.load_workbook('あいうえお表.xlsx')
```

```python
ws2 = wb2['Sheet1']
```

***

## ２．セル

■操作対象のセルの特定

```python
ws2['A1']
```

```python
ws2.cell(1, 1)
```

■セルの値の取得

```python
ws2['A1'].value
```

```python
ws2.cell(1, 1).value
```

■セルへの書き込み

```python
ws2.cell(1, 1).value
```

```python
ws2.cell(1, 1).value = '〇'
```

```python
ws2.cell(1, 1).value
```

■ワークブックの保存

```python
wb2.save('あいうえお表_修正版.xlsx')
```
