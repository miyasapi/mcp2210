## ７．条件分岐 ～条件に応じて、実行する処理を変える

自分が生まれた季節を変数「season」に代入し、seasonの値が「春」であれば「暖かい」、「夏」であれば「暑い」、  
「秋」であれば「涼しい」、「冬」であれば「寒い」と表示するプログラムを作成しましょう。  
もし、上記の季節以外の単語が入力された場合は、「その季節は登録されていません」と表示してください。

```python
# 自分が生まれた季節（ここでは秋）を変数「season」に代入
season = '秋'

# 変数「season」に入っている値が「春」かどうか、という条件分岐をつくる
if season == '春':
    # 条件を満たした際に、「暖かい」を表示させる
    print('暖かい')
elif season == '夏':
    print('暑い')
elif season == '秋':
    print('涼しい')
elif season == '冬':
    print('寒い')
# どの条件式にも一致しなかった際の処理を指定する
else:
    # 上記の季節以外の単語が入力された場合（どの条件も満たさなかった場合）に、
    # 「その季節は登録されていません」を表示させる
    print('その季節は登録されていません')
```

## ８．繰り返し処理 ～何度も同じ処理を行う

1から100までの数を表示してください。  
ただし、3の倍数の時は「Fizz」、5の倍数の時は「Buzz」、  
15の倍数の時は「FizzBuzz」と表示してください

```python
# for文とrange関数を使って、変数「i」に1から100までを順番に代入する
for i in range(1, 101):
    # もし変数「i」に入っている値が15の倍数であれば、「FizzBuzz」を表示させる
    # プログラムは上から順番に処理（順次構造）し、
    # 15の倍数は3の倍数、5の倍数の条件も満たすため、先頭に書く
    if i % 15 == 0:
        print('FizzBuzz')
    elif i % 3 == 0:
        print('Fizz')
    elif i % 5 == 0:
        print('Buzz')
    # どの条件も満たさなかった場合（3の倍数でも5の倍数でもなかった場合）は、
    # 変数「i」に入っている値をそのまま表示させる
    else:
        print(i)
```

## １０．関数 ～複数の処理をまとめて実行してくれる便利ツール１

以下のプログラムは、「あるクラスで実施したテストの、英語の○○を求める」プログラムです  
「○○」には何が入るでしょうか

```python
# 変数「english_scores」に、複数名の点数をリスト形式で代入する
english_scores = [54, 65, 71, 39, 87, 90, 81, 50, 67, 34, 64, 29, 98, 90, 52, 67]

# 後ほど使用する変数「total」に、初期値「0」を代入する
total = 0
# 変数「english_scores」から1つずつ値を取り出して、変数「score」に代入する
for score in english_scores:
    # 変数「total」に入っていた値と、変数「score」に入っている値を
    # 足し算し、改めて変数「total」に代入する
    # ⇒すべての処理を終えると、変数「total」には、すべての点数の合計が入る
    total = total + score
# 変数「english_scores」の要素数を求め、変数「student_num」に代入する
student_num = len(english_scores)
# 合計点が入っている変数「total」を、要素数（人数）が入っている変数「student_num」で
# 割り算を行い、結果を変数「x」に代入する
# ⇒「合計点÷人数」なので、平均点となる
x = total / student_num
print(x)
```

■解答

平均点
