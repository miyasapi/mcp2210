## ７．条件分岐 ～条件に応じて、実行する処理を変える

「if」のみ使用

```python
japanese_score = 50

if japanese_score >= 65:
    print('good')
```

「if」と「else」を使用

```python
if japanese_score >= 65:
    print('good')
else:
    print('soso')
```

「if」と「elif」と「else」を使用

```python
japanese_score = 65

if japanese_score >= 65:
    print('good')
elif japanese_score >= 30:
    print('soso')
else:
    print('bad')
```

「elif」を複数使用

```python
if japanese_score >= 90:
    print('very good')
elif japanese_score >= 65:
    print('good')
elif japanese_score >= 30:
    print('soso')
else:
    print('bad')
```

## ８．繰り返し処理 ～何度も同じ処理を行う

リスト「japanese_scores」を参照

```python
print(japanese_scores)
```

単純な条件分岐のプログラム

```python
japanese_score = 56

if japanese_score >= 60:
    print('可')
else:
    print('不可')
```

リストの各要素のテストの点数について、評価

```python
for score in japanese_scores:
    print(score)
    if score >= 60:
        print('可')
    else:
        print('不可')
    print('----------')
```

1から20までの数を表示

```python
for i in range(1,21):
    print(i)
```

0から99までの数を表示

```python
for i in range(100):
    print(i)
```

1から10の中で、5だけスキップ

```python
for i in range(1, 11):
    if i == 5:
        continue
    print(i)
```
